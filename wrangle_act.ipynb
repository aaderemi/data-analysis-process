## Quality
### `twitter_archive` table
- Incorrect names like 'a', 'an', 'the' etc., present in table.
- Some rows have more than one dog stage. As in index 889.
- Inconsistent numerator/denominator ratings.
- Tweet sources cannot be easily read from source column.
- Presence of actual names in some cases where `None` is filled for names in name column.
- Wrong data type for the timestamp column.
- Some of the tweets are retweets which have to be removed.
- Missing image links because some tweets are replies.
- Incorrect numerator rating and data type.

### `df_rt` table
- Missing number of retweets and likes for some tweets present in the `twitter_archive` table that have been deleted.

### `image_predictions` table
- There isn't a prediciton for every tweet in `twitter_archive`.

## Tidiness
- `df_rt`  and `image_predictions` tables should be part of the `twitter_archive` table.
- columns `doggo`, `floofer`, `puppo` and `pupper` in `twitter_archive` should be one column.
